```python
from math import sqrt

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as sct
import seaborn as sns
```

```python
#settings for matplotlib
%matplotlib inline 

from IPython.core.pylabtools import figsize

figsize(12, 8)

sns.set()
```

```python
#generates 1000 normal values of mean 10 and variance 9

norm = sct.norm.rvs(loc=10, scale=3, size=1000) #loc is the mean, scale is the standard deviation. X ~ N (10, 9).

sns.distplot(norm);
```

```python
#to check the length of the variable

len(norm)
```

```python
(norm.mean(), norm.var(), norm.std())
```

#### Normal Distribution

```python
#probability of values less than 8

sct.norm.cdf(8, loc=10, scale=3)
```

```python
#probability of values less than 10

sct.norm.cdf(10, loc=10, scale=3)
```

```python
#generates a CDF chart

x = np.linspace(0, 20, num=1000)

f = lambda x: sct.norm.cdf(x, loc=10, scale=3)

cdf = f(x)

sns.lineplot(x, cdf);
```

```python
#probability of numbers greater than 8

sct.norm.sf(8, loc=10, scale=3)
```

```python
#value for accumulated probability 80%

sct.norm.ppf(0.8, loc=10, scale=3)
```

```python
#quantile for cumulative probability greater than 25#

sct.norm.isf(0.25, loc=10, scale=3)
```

```python
#probability density

sct.norm.pdf(8, loc=10, scale=3)
```

```python
#generates the pdf chart

x = np.linspace(0, 20, num=1000)

f = lambda x: sct.norm.pdf(x, loc=10, scale=3)

cdf = f(x)

sns.lineplot(x, cdf);
```

#### Binominal Distribution

```python
binom_20_05 = sct.binom.rvs(20, 0.5, size=100000) # Mean 10, variance 5.
binom_20_07 = sct.binom.rvs(20, 0.7, size=100000) # Mean 14, variance 4.2.

sns.distplot(binom_20_05, bins=range(4, 18), kde=False, hist_kws={"alpha": 0.9});
sns.distplot(binom_20_07, bins=range(8, 22), kde=False, hist_kws={"alpha": 0.5});
```

Most of the functions available for the normal distribution that we present are also available for the binomial distribution:

<ul>
<li>mean(): $\mathbb{E}[X]$
<li>var(): $\text{Var}[X]$
<li>rvs(): $\vec{X}$ 
<li>cdf(): $P(X \leq x)$
<li>sf(): $P(X &gt; x)$ 
<li>ppf(): $x$ such that $P(X \leq x) = p$
<li>isf(): $x$ such that $P(X &gt; x) = p$
</ul>

The only significant change is in the pmf () ($ P (X = x) $) function, which was normally called pdf () ($ f (x) $).

##### What is the probability of obtaining a maximum of 3 heads (successes) in 20 launches of an addicted coin, which is $0.10 likely to be expensive?

```python
sum((sct.binom.pmf(i, 20, 0.10) for i in range(4)))
```

```python
sct.binom.cdf(3, 20, 0.10)
```

##### How likely are we to get at least 10 successes out of 20 releases of the same flawed coin above?

```python
1 - sum((sct.binom.pmf(i, 20, 0.10) for i in range(10)))
```

```python
1 - sct.binom.cdf(9, 20, 0.10)
```

```python
sct.binom.sf(9, 20, 0.10)
```

##### How likely are we to get at least 50 successes out of 200 launches of an addicted currency that has a probability of success of $0.3?

```python
1 - sct.binom.cdf(49, 200, 0.30)
```

```python
1 - sct.norm.cdf(50, 60, sqrt(42)) # Y ~ N(60, 42).
```

```python
1 - sct.binom.cdf(9, 20, 0.10)
```
